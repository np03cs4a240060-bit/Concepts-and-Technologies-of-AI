Solving the linear equation

```python
import numpy as np

# Coefficient matrix A
A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3,  1, -1]
])

# Constant matrix B
B = np.array([-1, -3, 9])

# Solving using Inverse Method: X = A^(-1) * B
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)

print("Solution using Inverse Method:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])

# Solving using np.linalg.solve (more efficient and recommended)
X_solve = np.linalg.solve(A, B)

print("\nSolution using np.linalg.solve():")
print("x =", X_solve[0])
print("y =", X_solve[1])
print("z =", X_solve[2])
```
